```python
order = [int(x) for x in data.pop(0).split(",")]
boards = []
for board in data:
    board = [int(x) for x in board.split()]
    boards.append(board)
```

```python
def is_bingo(board):
    # check rows
    for i in range(5):
        cand = True
        for j in range(5):
            if board[5*i+j] != -1:
                cand = False
        if cand:
            return True
    # check columns
    for i in range(5):
        cand = True
        for j in range(5):
            if board[i+5*j] != -1:
                cand = False
        if cand:
            return True
    return False
```

```python
from copy import copy

fresh_boards = copy(boards)

for num in order:
    bingo = False
    fresh_boards = [[-1 if x==num else x for x in board] for board in fresh_boards]
    for board in fresh_boards:
        if is_bingo(board):
            bingo = True
            winner_board = board
            winner_number = num
            break
    if bingo:
        break

print(sum([x for x in winner_board if x != -1]) * winner_number)
```
