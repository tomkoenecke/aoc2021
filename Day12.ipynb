```python
connections = []
for line in data:
    connections.append(tuple(line.split("-")))
connections
```

```python
def all_neighbors(connections, me):
    neighbors = []
    for start, end in connections:
        if me == start:
            neighbors.append(end)
        if me == end:
            neighbors.append(start)
    return neighbors
```

```python
from copy import copy
count = 0
paths = [["start"]]
while len(paths) > 0:
    path = paths.pop(0)
    latest = path[-1]
    if latest == "end":
        count += 1
        continue
    neighbors = all_neighbors(connections, latest)
    for neighbor in neighbors:
        if neighbor.lower() in path:
            continue
        newpath = copy(path)
        newpath.append(neighbor)
        paths.append(newpath)
count
```

```python
def contains_double_small_cave(path):
    for element in path:
        if element.lower() == element:
            count = 0
            for element2 in path:
                if element == element2:
                    count += 1
            if count > 1:
                return True
    return False
```

```python
count = 0
paths = [["start"]]
while len(paths) > 0:
    path = paths.pop(0)
    latest = path[-1]
    if latest == "end":
        count += 1
        continue
    neighbors = all_neighbors(connections, latest)
    for neighbor in neighbors:
        if neighbor == "start":
            continue
        if neighbor.lower() in path:
            # only skip, if there is already a double small cave in path!
            if contains_double_small_cave(path):
                continue
        newpath = copy(path)
        newpath.append(neighbor)
        paths.append(newpath)
count
```
