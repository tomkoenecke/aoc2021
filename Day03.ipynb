```python
def bin_to_dec(bin):
    bin = list(bin)
    bin.reverse()
    factor = 1
    dec = 0
    for letter in bin:
        dec += factor*int(letter)
        factor *= 2
    return dec
```

```python
def complementary_bin(bin):
    out = ""
    for bit in bin:
        if bit == "1":
            out += "0"
        else:
            out += "1"
    return out
```
