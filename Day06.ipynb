```python
def spawn(days_left):
    if days_left < -1:
        return 0
    if days_left <= 7:
        return 1
    num = 1
    while days_left > 6:
        days_left -= 7
        # print("Fish Spawned at", days_left, "days_left")
        num += spawn(days_left - 2)
    return num
```
