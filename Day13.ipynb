```python
dots, folds = data.split("\n\n")
dots = dots.splitlines()
dots = [[int(x) for x in dot.split(",")] for dot in dots]
folds = [x.split()[2].split("=") for x in folds.splitlines()]
```

```python
maxx = 0
maxy = 0
for dot in dots:
    x, y = dot
    maxx = max(x + 1, maxx)
    maxy = max(y + 1, maxy)
print(maxx, maxy)
```

# Part 1

```python
smallestx = maxx
smallesty = maxy
for dir, fold in folds:
    fold = int(fold)
    n = len(dots)
    for _ in range(n):
        dot = dots.pop(0)
        if dir == "x":
            x, y = dot
            if x > fold:
                x = fold - (x-fold)
            dots.append([x, y])
            smallestx = fold

        elif dir == "y":
            x, y = dot
            if y > fold:
                y = fold - (y - fold)
            dots.append([x, y])
            smallesty = fold
    break
```

```python
# Initialize grid
grid = [["_" for x in range(smallestx)] for y in range(smallesty)]

# Draw dots on grid
for dot in dots:
    x, y = dot
    grid[y][x] = "#"
```
