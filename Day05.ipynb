```python
data = """0,9 -> 5,9
8,0 -> 0,8
9,4 -> 3,4
2,2 -> 2,1
7,0 -> 7,4
6,4 -> 2,0
0,9 -> 2,9
3,4 -> 1,4
0,0 -> 8,8
5,5 -> 8,2""".splitlines()
```

```python
# determine grid size and collect lines from data
lines = []
grid_size = 0
for line in data:
    start, _ , end = line.split()
    start = [int(x) for x in start.split(",")]
    end = [int(x) for x in end.split(",")]
    grid_size = max(max(start), max(end), grid_size)
    lines.append([start, end])
```

```python
# Part 1
grid = [[0 for _ in range(grid_size + 1)] for _ in range(grid_size + 1)]

for line in lines:
    start, end = line
    # vertical line
    if start[0] == end[0]:
        col = start[0]
        if start[1] <= end[1]:
            for i in range(start[1], end[1] + 1):
                grid[i][col] += 1
        elif start[1] > end[1]:
            for i in range(end[1], start[1] + 1):
                grid[i][col] += 1
    # horizontal line
    if start[1] == end[1]:
        row = start[1]
        if start[0] <= end[0]:
            for i in range(start[0], end[0] + 1):
                grid[row][i] += 1
        elif start[0] > end[0]:
            for i in range(end[0], start[0] + 1):
                grid[row][i] += 1

# count all cells >2
count = 0
for line in grid:
    for element in line:
        if element >= 2:
            count += 1
print(count)
```

```python
# Part 2
grid = [[0 for _ in range(grid_size + 1)] for _ in range(grid_size + 1)]

for line in lines:
    start, end = line
    row, col = start
    
    while row != end[0] or col != end[1]:
        grid[col][row] += 1
        if row < end[0]:
            row += 1
        elif row == end[0]:
            pass
        else:
            row -= 1

        if col < end[1]:
            col += 1
        elif col == end[1]:
            pass
        else:
            col -= 1
    grid[col][row] += 1

# count all cells >2
count = 0
for line in grid:
    for element in line:
        if element >= 2:
            count += 1
print(count)
```
