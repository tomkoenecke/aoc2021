```python
def greater_nine(grid):
    out = []
    for y, line in enumerate(grid):
        for x, num in enumerate(line):
            if num >= 10:
                out.append((x,y))
    return out
```

# Part 1

```python
# create grid from data
grid = [[int(x) for x in line] for line in data]

flash_count = 0
for _ in range(100):
    # energy level of each octopus increases by 1
    grid = [[x + 1 for x in line] for line in grid]

    # all 9s flash
    flashed = []
    tens_exist = True
    while len(greater_nine(grid)) > 0:
        tens = greater_nine(grid)
        for coords in tens:
            x, y = coords
            flashed.append((x,y))
            flash_count += 1
            grid[y][x] = -20
            for dy in [-1, 0, 1]:
                for dx in [-1, 0, 1]:
                    newx = x + dx
                    newy = y + dy
                    if newy >= 0 and newx >= 0 and newy < len(grid) and newx < len(grid[0]):
                        grid[newy][newx] += 1

    # set all that flashed to 0
    for x, y in flashed:
        grid[y][x] = 0
        
flash_count
```
