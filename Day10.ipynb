```python
data = """[({(<(())[]>[[{[]{<()<>>
[(()[<>])]({[<{<<[]>>(
{([(<{}[<>[]}>{[]{[(<()>
(((({<>}<{<{<>}{[]{[]{}
[[<[([]))<([[{}[[()]]]
[{[{({}]{}}([{[{{{}}([]
{<[[]]>}<{[{[{[]{()[[[]
[<(<(<(<{}))><([]([]()
<{([([[(<>()){}]>(<<{{
<{([{{}}[<[[[<>{}]]]>[]]""".splitlines()
```

```python
openers = ["(", "[", "{", "<"]
closers = [")", "]", "}", ">"]
rel = {a:b for a,b in zip(openers, closers)}
```

# Part 1

```python
scores = {")":3, "]":57, "}":1197, ">":25137}
scores
```

```python
total = 0
incomps = []
for line in data:
    corrupt = False
    mem = []
    for char in line:
        if char in openers:
            mem.append(char)
        elif char in closers:
            if char == rel[mem[-1]]:
                mem.pop(-1)
            else:
                total += scores[char]
                corrupt = True
                break
    # preparation for part 2
    if len(mem) != 0 and not corrupt:
        incomps.append(line)
        
total
```

# Part 2

```python
scores = {")":1, "]":2, "}":3, ">":4}
```

```python
all_scores = []
for line in incomps:
    mem = []
    for char in line:
        if char in openers:
            mem.append(char)
        elif char in closers:
            if char == rel[mem[-1]]:
                mem.pop(-1)
            else:
                print("This line should not be here, because it is corrupt", line)
    appendum = ""
    for m in mem:
        appendum = rel[m] + appendum
    
    # scoring
    score = 0
    for char in appendum:
        score *= 5
        score += scores[char]
    all_scores.append(score)
```

```python
# find middle score
sorted(all_scores)[int(len(all_scores)/2)]
```
